```python
import os
def read_data(datadir):
    files = [x[:-1] for x in  open(os.path.join(datadir, "datafiles.txt")).readlines()]
    return np.concatenate([np.load(os.path.join(datadir, fn))['arr_0'] for fn in files])
                 
data = read_data("./data/")
                 
contourf(data['u'][:,1,:], 12, cmap='bwr')
colorbar()
```
